### Q31. Label numbers Even/Odd.

```python
numbers = [10, 15, 22, 33, 40]

result = []

for n in numbers:
    if n % 2 == 0:
        result.append((n, "Even"))
    else:
        result.append((n, "Odd"))

print(result)

```
